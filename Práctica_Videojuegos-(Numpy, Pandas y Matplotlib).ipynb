```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\apico\Downloads\vgsales.csv")
df.head()
```

**Pregunta 1:** ¿Cuál es la plataforma más popular en términos de ventas de videojuegos?

Podemos abordar este problema trazando las ventas totales de cada plataforma. Esto nos permitiría ver rápidamente qué plataforma tiene las ventas más altas, lo que podría ser útil para los desarrolladores de juegos al decidir para qué plataformas desarrollar sus juegos.

```python
# Plataforma mas popular
# Plataforma eje x ventas eje y

fig = plt.figure(figsize=(10,6))

df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).plot(kind='bar')

plt.xlabel('Plataforma')
plt.ylabel('Ventas Globales M')
plt.title('Ventas Globales por Plataforma')

plt.show()
```

**Pregunta 2:** ¿Cómo han cambiado las ventas de videojuegos a lo largo del tiempo?

Para esto, podríamos trazar las ventas totales de videojuegos por año. Esto podría mostrar tendencias interesantes, como si las ventas están aumentando o disminuyendo con el tiempo.

```python
# Ventas según año
# Año eje x ventas eje y
# En este caso no hace falta poner el sort_values porque los numeros los ordena por defecto
fig = plt.figure(figsize=(10,6))

df.groupby('Year')['Global_Sales'].sum().plot(kind='line')

plt.xlabel('Año')
plt.ylabel('Ventas Globales M')
plt.title('Ventas Globales por Año')

plt.show()
```

**Pregunta 3:** ¿Cuáles son los géneros de videojuegos más populares en diferentes regiones?

Podríamos abordar este problema creando gráficos de barras apiladas para cada región, mostrando las ventas de cada género. Esto podría revelar diferencias regionales en las preferencias de género.

```python
# Plataforma mas popular
# Generos eje x ventas eje y

regiones = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']

for regions in regiones:
    fig = plt.figure(figsize=(10,6))
    df.groupby('Genre')[regions].sum().sort_values(ascending= False).plot(kind= 'bar')
    plt.title(f'Ventas por genero en {regions}')
    plt.ylabel('Ventas Globales M')
    plt.show()

```

**Pregunta 4:** ¿Cuáles son los editores de videojuegos más exitosos en términos de ventas globales?

Para responder a esta pregunta, podríamos trazar las ventas globales de cada editor. Esto podría ser útil para los inversores que buscan invertir en empresas de videojuegos.

```python

plt.figure(figsize=(10,6))

df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10).plot(kind='bar')

plt.ylabel('Ventas Globales M')
plt.title('Ventas Globales por Editores')
plt.show()
```
