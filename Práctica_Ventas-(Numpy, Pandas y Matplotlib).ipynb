## Ejercicio 1: lee el beneficio total de todos los meses y muéstralo mediante un diagrama de líneas

Datos de ganancias totales proporcionados para cada mes. El diagrama de líneas generado debe incluir las siguientes propiedades: –

    Nombre de etiqueta X = Número de mes
    Nombre de la etiqueta Y = Beneficio total

El gráfico de trazado de líneas debería tener este aspecto.

```python
# Mostrar diagrama de lineas con beneficio total por mes
# x = numero de mes
# y = beneficio total

monthList = df['month_number'].tolist() # x
profitList= df['total_profit'].tolist() # y

# Creamos lienzo con un tamaño mas o menos igual a la muestra
fig = plt.figure(figsize= (4,3))

# Introducimos los valores en el gráfico
plt.plot(monthList,profitList)

# Le damos forma al gráfico
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.title('Company profit per month')
plt.xticks(range(1, 13, 1))                # Creamos rango para eje x
plt.yticks(range(100000, 600000, 100000))  # Creamos rango para eje y

plt.show()
```

## Ejercicio 3: lee todos los datos de ventas de productos y muéstralos usando un gráfico multilínea

Muestre el número de unidades vendidas por mes para cada producto utilizando gráficos de líneas múltiples. (es decir, trama separada para cada producto).

El gráfico debería verse así.

```python
# Creamos lienzo en blanco con tamaño especificado

fig = plt.figure(figsize=(4,3))

# Le damos forma al gráfico
plt.title('Sales data')
plt.xlabel('Month number')
plt.ylabel('Sales unit in number')
plt.xticks(range(1, 13, 1))                                             # Creamos rango para eje x
plt.ylim(0,18000)                                                       # Creamos límite del eje y
etiquetas = ['',1000,2000,4000,6000,8000,10000,12000,14000,16000,18000] # Creamos etiquetas para cada escalón del eje y
plt.yticks(range(0,22000,2000),etiquetas)                               # Aplicamos etiquetas para cada escalón del eje y


# Definimos variables del gráfico

# Lista para eje x

monthList = df['month_number'].tolist() 

# Lista para eje y

# Creamos diccionario para iterar sobre los valores y sus colores

productos= {'facecream': 'blue', 'facewash': 'orange', 'toothpaste':'green' , 'bathingsoap': 'red', 'shampoo': 'purple', 'moisturizer': 'brown'}

for clave,valor in productos.items():
   
    new_line = df[clave].tolist() # Con cada iteración se crea una lista que corresponde a cada tipo de producto
    plt.plot(monthList,new_line, color = valor, lw=2, ls='-', marker='o', markersize=4,label= clave +' ' + "sales data") # A cada iteración se le asigna un color y su leyenda.
    plt.legend(loc="upper left", prop = {'size':7});                               

plt.show()


```

## Ejercicio 8: Calcula los datos de ventas totales del último año para cada producto y muéstralos usando un gráfico circular (pie chart)

Nota: En la visualización del gráfico circular Número de unidades vendidas por año para cada producto en porcentaje.

El gráfico circular debería verse así.

```python


# Creamos lienzo en blanco con tamaño especificado

fig= plt.figure(figsize=(4,3))


# Creamos diccionario y variables para iterar sobre los valores y sus colores

productos= {'facecream': 'blue', 'facewash': 'orange', 'toothpaste':'green' , 'bathingsoap': 'red', 'shampoo': 'purple', 'moisturizer': 'brown'}

totales_ventas_productos = []
lista_productos = []
lista_colores = []

for clave,valor in productos.items():

    new_line = df[clave].sum()                # Variable que suma los totales de cada producto
    totales_ventas_productos.append(new_line) 
    lista_productos.append(clave)
    lista_colores.append(valor)
    
# Creamos el grafico circular con lo datos obtenidos en el for

plt.pie(totales_ventas_productos, labels= lista_productos, colors= lista_colores , autopct='%1.1f%%')
plt.title('Sales data')
plt.legend(loc=6, prop = {'size':6}); # Imprimimos la leyenda a la izquierda para que no tape los nombres

plt.axis('equal')
plt.show()

```

## Ejercicio 10: Lee todos los datos de ventas de productos y muéstralos usando el diagrama de pila

El diagrama de pila debería verse así.

```python
# Creamos lienzo en blanco con tamaño especificado

fig = plt.figure(figsize=(4,3))

# Definimos variables del gráfico

# Lista para eje x

monthList = df['month_number'].tolist() 

# Listas para eje y

facecream = df['facecream'].tolist()
facewash = df['facewash'].tolist()
toothpaste = df['toothpaste'].tolist()
bathingsoap = df['bathingsoap'].tolist()
shampoo = df['shampoo'].tolist()
moisturizer = df['moisturizer'].tolist()

# Introducimos valores y leyenda en el gráfico

etiquetas = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
colores = ['darkmagenta','darkturquoise','red','black', 'forestgreen','darkkhaki']

plt.stackplot(monthList, facecream, facewash, toothpaste, bathingsoap, shampoo, moisturizer, labels=etiquetas, colors= colores)
plt.legend(loc="upper left", prop = {'size':7});                               


# Le damos forma al gráfico
plt.title('Sales data')
plt.xlabel('Month number')
plt.ylabel('Sales unit in number')
plt.xticks(range(0, 13, 2))                # Creamos rango para eje x
plt.ylim(0,33000)                          # Creamos límite del eje y
plt.yticks(range(0,33000,5000))            # Aplicamos rango valores para el eje y

    
plt.show()
```
