```python
# Concatenamos las tablas para tenerlo todo en una misma tabla
emisiones = pd.concat([emisiones2016,emisiones2017,emisiones2018,emisiones2019], axis= 0)
emisiones = pd.DataFrame(emisiones)
emisiones
```

### 2. Filtrar las columnas del DataFrame para quedarse con las columnas ESTACION, MAGNITUD, AÑO, MES y las correspondientes a los días D01, D02, etc.

```python
# Filtramos las columnas solicitadas y eliminamos las innecesarias
emisiones_1 = emisiones.drop(['PROVINCIA','MUNICIPIO','PUNTO_MUESTREO'], axis= 1)
filtrado_v = emisiones_1.iloc[:,5:67:2]
emisiones_2 = emisiones_1.drop(filtrado_v, axis= 1)
emisiones_2 
```

### 3. Reestructurar el DataFrame para que los valores de los contaminantes de las columnas de los días aparezcan en una única columna.

```python
# Mediante la función melt, convertimos el ancho en largo para tener todo en una columna
emisiones_3 = emisiones_2.melt(id_vars=['ESTACION', 'MAGNITUD','ANO', 'MES'], var_name='DIA', value_name='VALOR')
# Eliminimos la D de la columna día
emisiones_3['DIA'] = emisiones_3['DIA'].str.lstrip('D') 
emisiones_3
```

### 4. Añadir una columna con la fecha a partir de la concatenación del año, el mes y el día (usar el módulo datetime).

```python
# Creamos una función que concatene las columnas
def concatenar_fecha(row):
    date_str = f"{row['ANO']}-{row['MES']}-{row['DIA']}" # Filtramos y pasamos a string la fecha
    try:
        date = datetime.strptime(date_str, "%Y-%m-%d") # Pasamos el sting a formato fecha
        date = date.strftime('%Y-%m-%d') # Eliminamos las horas, minutos y segundos 
    except ValueError:
        date = "NaT"  # Cuando la fila no tiene valores, sale un error en la fecha. Lo eliminamos con esta función
    return date

# Creamos la nueva columna y aplicamos la función

emisiones_3['FECHA']= emisiones_3.apply(concatenar_fecha,axis=1)
emisiones_3
```

### 5. Eliminar las filas con fechas no válidas (utilizar la función isnat del módulo numpy) y ordenar el DataFrame por estaciones contaminantes y fecha.

```python

# Convertimos la columna 'FECHA' a datetime

emisiones_3['FECHA'] = pd.to_datetime(emisiones_3['FECHA'])

# Filtramos las filas con fechas nulas

nulas = np.isnat(emisiones_3['FECHA'])

# Eliminar las filas usando la inversa(elimina las malas y solo imprime las buenas)
emisiones_4 = emisiones_3[~nulas]

# Ordenamos DF por estaciones contaminantes y fecha

emisiones_4 = emisiones_4.sort_values(['ESTACION','FECHA'])

emisiones_4
```

### 6. Mostrar por pantalla las estaciones y los contaminantes disponibles en el DataFrame.

```python

# Mostrar los valores únicos de la columna 'ESTACION' y 'MAGNITUD'
estaciones = emisiones_4['ESTACION'].unique()
contaminantes = emisiones_4['MAGNITUD'].unique()

# Mostrar los valores únicos
print("Las estaciones disponibles son",estaciones)
print("Los contaminantes son",contaminantes)
```

### 7. Crear una función que reciba una estación, un contaminante y un rango de fechas y devuelva una serie con las emisiones del contaminante dado en la estación y rango de fechas dado.

```python
estacion = 11
contaminante = 12
fecha_inicio = '2016-01-01'
fecha_fin = '2016-08-01'

def filtro_emisiones(estacion,contaminante,fecha_inicio,fecha_fin):
    
    # Filtramos los valores dados en las columnas correspondientes
    filtro_emi = emisiones_4[emisiones_4['ESTACION'].isin([estacion])] 
    filtro_con = emisiones_4[emisiones_4['MAGNITUD'].isin([contaminante])]
    filtro_emi_con = filtro_emi.merge(filtro_con) # Comprobamos los elementos en común entre ESTACION y MAGNITUD


    filtro_fecha_inicio = emisiones_4[emisiones_4['FECHA'] >= fecha_inicio]
    filtro_fecha_fin = emisiones_4[emisiones_4['FECHA'] <= fecha_fin]

    filtro_fechas = filtro_fecha_inicio.merge(filtro_fecha_fin) # Comprobamos las fechas en común entre la de inicio y fin
    
    filtro_total = filtro_emi_con.merge(filtro_fechas) # Comprobamos elementos en comun entre todas las condiciones
    
    
    return filtro_total['VALOR'] # Filtramos solo la columna Valor para que nos salga el resultado solicitado


emisiones_filtradas = filtro_emisiones(estacion, contaminante, fecha_inicio, fecha_fin)

emisiones_filtradas
```

### 8. Mostrar un resumen descriptivo (mínimo, máximo, media, etc.) para cada contaminante.

```python

# Filtramos las columnas que vamos a utilizar

emisiones_5 = emisiones_4[['MAGNITUD','VALOR']]

# Agrupamos por contaminante y aplicamos el resumen descriptivo al valor

resumen_descriptivo = emisiones_5.groupby('MAGNITUD')['VALOR'].describe()

resumen_descriptivo
```

### 10. Crear una función que reciba una estación y un contaminante y devuelva un resumen descriptivo de las emisiones del contaminante indicado en la estación indicada.

```python
estacion = 11
contaminante = 12

def resumen_descriptivo_emisiones(estacion,contaminante):
    
    # Filtramos los valores dados en las columnas correspondientes
    filtro_emi = emisiones_4[emisiones_4['ESTACION'].isin([estacion])] 
    filtro_con = emisiones_4[emisiones_4['MAGNITUD'].isin([contaminante])]
    filtro_emi_con = filtro_emi.merge(filtro_con) # Comprobamos los elementos en común entre ESTACION y MAGNITUD
    filtro_total = filtro_emi_con['VALOR'].describe() # Aplicamos resumen descriptivo a la columna VALOR
    
    return filtro_total

resumen_descriptivo_filtrado = resumen_descriptivo_emisiones(estacion,contaminante)

resumen_descriptivo_filtrado
```

### 11. Crear una función que devuelva las emisiones medias mensuales de un contaminante y un año dados para todos las estaciones.

```python
contaminante = 12

def emisiones_medias_mensuales(contaminante):
    # Filtramos los valores dados en las columnas correspondientes   
    filtro_con = emisiones_4[emisiones_4['MAGNITUD'].isin([contaminante])]    
    media_emisiones = filtro_con.groupby(['ESTACION','ANO','MES'])['VALOR'].mean() # Agrupamos resultados y calculamos la media
    return media_emisiones

emisiones_mensuales = emisiones_medias_mensuales(contaminante)
emisiones_mensuales
```
